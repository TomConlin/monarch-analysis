```python
import requests
import copy

SOLR_URL = 'https://solr-dev.monarchinitiative.org/solr/golr/select'
SCIGRAPH_URL = 'https://scigraph-ontology-dev.monarchinitiative.org/scigraph'

filters = [
    'object_category:"disease"',
    'subject_category:"gene"',
    'qualifier:"inferred through variant"',
    'subject_taxon: "{0}"'.format('NCBITaxon:9606')
]

params = {
    'wt': 'json',
    'rows': 1000,
    'start': 0,
    'q': '*:*',
    'fq': filters,
    'fl': 'subject, subject_label, object, object_label, relation, is_defined_by'
}

# Copy paste function (need to put somewhere)
def get_solr_results(solr, params):
    solr_params = copy.deepcopy(params)
    resultCount = solr_params['rows']
    while solr_params['start'] < resultCount:
        solr_request = requests.get(solr, params=solr_params)
        response = solr_request.json()
        resultCount = response['response']['numFound']
        solr_params['start'] += solr_params['rows']
        for doc in response['response']['docs']:
            yield doc

count = 0
for doc in get_solr_results(SOLR_URL, params):
    if 'https://data.monarchinitiative.org/ttl/omim.ttl' in doc['is_defined_by'] and \
            'https://data.monarchinitiative.org/ttl/clinvar.ttl' not in doc['is_defined_by']:
        print(doc['subject'])
        count += 1
count
            
```

```python
def check_disease_position(disease:str) -> bool:
    isLeafNode = False
    scigraph_service = SCIGRAPH_URL + "/graph/neighbors"
    params = {
        "id": disease,
        "relationshipType": "subClassOf",
        "direction": "INCOMING",
        "depth": 1
    }

    req = requests.get(scigraph_service, params=params)
    results = req.json()

    if len(results["edges"]) == 0:
        isLeafNode = True
        
    return isLeafNode

leafCount = 0
groupCount = 0
for doc in get_solr_results(SOLR_URL, params):
    if 'https://data.monarchinitiative.org/ttl/orphanet.ttl' in doc['is_defined_by']:
        isLeafNode = check_disease_position(doc['object'])
        if isLeafNode:
            leafCount += 1
        else:
            groupCount +=1
print("{} associations to leaf nodes".format(leafCount))
print("{} associations to disease groups".format(groupCount))
```
